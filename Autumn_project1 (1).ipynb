## EDA including visualization

### Plot 1:

Monthly Revenue Over Time  

Visualizing monthly revenue helps us understand sales trends across the year.  
Are there peaks during certain months (like festivals or holidays)? Lets find out.

```python
# Ensure Date column is in datetime format
df_sales['Date'] = pd.to_datetime(df_sales['Date'])

# Group by Month-Year and calculate total revenue
monthly_revenue = df_sales.groupby(df_sales['Date'].dt.to_period('M'))['Total Amount'].sum()

# Convert PeriodIndex to Timestamp for plotting
monthly_revenue.index = monthly_revenue.index.to_timestamp()

# ---- Line Chart ----
plt.figure(figsize=(10,6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', linewidth=2)
plt.title("Monthly Revenue Over Time", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# ---- Bar Chart ----
plt.figure(figsize=(10,6))
plt.bar(monthly_revenue.index, monthly_revenue.values, width=20)  # width=20 for month spacing
plt.title("Monthly Revenue (Bar Chart)", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.show()
```

Plot the weekly revenue trend instead of monthly. What differences do you notice?


```python
# Group by Week-Year and calculate total revenue
weekly_revenue = df_sales.groupby(df_sales['Date'].dt.to_period('W'))['Total Amount'].sum()

# Convert PeriodIndex to Timestamp for plotting
weekly_revenue.index = weekly_revenue.index.to_timestamp()

# Plot weekly revenue
plt.figure(figsize=(12, 6))
plt.plot(weekly_revenue.index, weekly_revenue.values)
plt.title("Weekly Revenue Over Time")
plt.xlabel("Week")
plt.ylabel("Revenue")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###Plot 2

Now well try to visualize the overall sales over all the product categories and infer about what product category is best selling.

```python
# Group by Product Category and sum the revenue
category_revenue = df_sales.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(8,6))
category_revenue.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Best-Selling Categories by Revenue", fontsize=14)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Total Revenue", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.show()
```

### Question:  
What if we want to visualize the **top 5 customers** by total spending?  
Try writing the code to generate a bar chart!  

```python
# Group by Customer ID and sum the total amount spent
customer_spending = df_sales.groupby('Customer ID')['Total Amount'].sum()

# Get the top 5 customers by total spending
top_5_customers = customer_spending.nlargest(5)

# Plot bar chart of top 5 customers
plt.figure(figsize=(10, 6))
top_5_customers.plot(kind='bar', color='teal', edgecolor='black')

plt.title("Top 5 Customers by Total Spending", fontsize=14)
plt.xlabel("Customer ID", fontsize=12)
plt.ylabel("Total Spending", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

Calculate the average order value (AOV) per customer and plot them with a suitable diagram. Which customers have unusually high AOVs?

```python
# Calculate Average Order Value (AOV) per customer
# Group by Customer ID and calculate the mean of 'Total Amount'
customer_aov = df_sales.groupby('Customer ID')['Total Amount'].mean()

# Display the first few values
print("Average Order Value (AOV) per Customer:")
print(customer_aov.head())

# Plotting the distribution of AOV using a histogram
plt.figure(figsize=(10, 6))
plt.hist(customer_aov, bins=50, edgecolor='black')
plt.title("Distribution of Average Order Value (AOV) per Customer", fontsize=14)
plt.xlabel("Average Order Value (AOV)", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.grid(axis='y', linestyle="--", alpha=0.6)
plt.show()

# Plotting the distribution of AOV using a box plot to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=customer_aov)
plt.title("Box Plot of Average Order Value (AOV) per Customer", fontsize=14)
plt.xlabel("Average Order Value (AOV)", fontsize=12)
plt.grid(axis='x', linestyle="--", alpha=0.6)
plt.show()

# Identifying customers with unusually high AOVs (outliers from the box plot)
# We can use the quantile method to find a threshold for high AOV
q3 = customer_aov.quantile(0.75)
iqr = customer_aov.quantile(0.75) - customer_aov.quantile(0.25)
upper_bound = q3 + 1.5 * iqr

unusually_high_aov_customers = customer_aov[customer_aov > upper_bound]

print("\nCustomers with Unusually High Average Order Value (AOV):")
print(unusually_high_aov_customers)
```

##Task:
Plot a cumulative revenue line chart. How does this help in understanding long-term sales growth?

```python
# Ensure Date column is in datetime format
df_sales['Date'] = pd.to_datetime(df_sales['Date'])

# Sort by date to ensure correct cumulative sum
df_sales_sorted = df_sales.sort_values(by='Date')

# Calculate cumulative revenue
df_sales_sorted['Cumulative Revenue'] = df_sales_sorted['Total Amount'].cumsum()

# Group by Month-Year to plot cumulative monthly revenue
cumulative_monthly_revenue = df_sales_sorted.groupby(df_sales_sorted['Date'].dt.to_period('M'))['Cumulative Revenue'].max()

# Convert PeriodIndex to Timestamp for plotting
cumulative_monthly_revenue.index = cumulative_monthly_revenue.index.to_timestamp()

# Plot cumulative revenue
plt.figure(figsize=(10, 6))
plt.plot(cumulative_monthly_revenue.index, cumulative_monthly_revenue.values, marker='o', linestyle='-')
plt.title("Cumulative Revenue Over Time", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Cumulative Revenue", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Explanation of how cumulative revenue helps understand long-term sales growth
print("\nHow the Cumulative Revenue Line Chart Helps Understand Long-Term Sales Growth:")
print("A cumulative revenue line chart shows the running total of sales over time.")
print("This type of chart is useful for:")
print("- Visualizing the overall growth trend: A consistently upward slope indicates steady sales growth.")
print("- Identifying milestones: You can see when certain revenue targets were reached.")
print("- Comparing growth rates over different periods: The steepness of the curve shows how quickly revenue is accumulating.")
print("- Understanding the total revenue generated up to any given point in time.")
print("In essence, it provides a clear picture of the business's overall financial performance and its trajectory over the long term.")
```
